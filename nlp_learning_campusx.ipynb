```python
class perform_nlp:

  def __init__(self):
    self.generes_df=pd.DataFrame()
    self.movies_df=pd.DataFrame()

  def pull_movie_list(self):
    _movie_list=[]
    #for _r in range(1,494+1):
    for _r in range(1,50):
      _movie_url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page="+str(_r)
      movies=requests.get(_movie_url).text
      _movie_list=_movie_list + json.loads(BeautifulSoup(movies).text)['results']
    return _movie_list

  def pull_genres(self):
    _genere_url='https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US'
    generes=requests.get(_genere_url).text
    return json.loads(BeautifulSoup(generes).text)['genres']

  def join_both_data(self):
    _columns=['original_title','overview','generes']
    self.movies_df=pd.DataFrame(self.pull_movie_list())
    self.generes_df=pd.DataFrame(self.pull_genres())
    self.movies_df['generes']=self.movies_df.apply(lambda x:', '.join(self.generes_df[self.generes_df['id'].isin(x['genre_ids'])]['name']),axis=1)
    _movies=self.movies_df.loc[:,_columns]
    return _movies

  def stemming(self,_text_arr):
    return [ps.stem(_stem_words) for _stem_words in _text_arr]

  def perform_txt_processing(self,_data_for_editing):
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.lowercase_sentence(x))
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.remove_special_characters(x))
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.tokenize_word(x))
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.remove_stopwords(x))
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.check_spelling(x))
    _data_for_editing['overview']=_data_for_editing['overview'].apply(lambda x:self.stemming(x))
    return _data_for_editing

  def lowercase_sentence(self,_text):
    return _text.lower().strip()

  def remove_special_characters(self,_text):
    _compiled_regex=re.compile('[a-zA-Z\.0-9 ]*')
    return ''.join(_compiled_regex.findall(_text))

  def tokenize_word(self,_text):
    return word_tokenize(_text)

  def check_spelling(self,_text_arr):
    return [english.correction(_word) for _word in _text_arr]

  def remove_stopwords(self,_text_arr):
    return [_word for _word in _text_arr if _word not in stopwords.words('english')]

  def main(self):
    _movies=self.join_both_data()
    _movies=self.perform_txt_processing(_movies)
    return _movies

_nlp=perform_nlp()
_movies=_nlp.main()
```
