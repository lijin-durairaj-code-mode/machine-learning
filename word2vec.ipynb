```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#gensim & nltk
import os
import gensim
from gensim import corpora
from gensim.parsing.preprocessing import remove_stopwords
from gensim.utils import simple_preprocess
from nltk import sent_tokenize

#sklearn
from sklearn.decomposition import PCA

#configuration
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

#plot
import plotly.express as px
```

```python
processed_data=[]
for _lists in os.listdir('./sample_data'):
  if(_lists.endswith('.txt')):
    path=os.path.join('./sample_data',_lists)
    with open(path,'r',errors='ignore') as _txt:
      _tokenized_d=sent_tokenize(remove_stopwords(_txt.read()))
      for _token in _tokenized_d:
        processed_data.append(simple_preprocess(_token))
```

```python
len(processed_data)
```

```python
model=gensim.models.Word2Vec(window=10, min_count=3)
model.build_vocab(processed_data)
model.train(processed_data,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
y=model.wv.index2word
```

```python
len(y)
```

```python
pca=PCA(n_components=3)
_plot=pca.fit_transform(model.wv.vectors)
fig=px.scatter_3d(_plot,x=0,y=1,z=2,color=y)
fig.show()
```
